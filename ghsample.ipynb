```python
#sample

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Google Colab での表示設定
%matplotlib inline

# ランダムなデータの生成 (例: 身長と体重のデータ)
np.random.seed(42)
num_samples = 100

height = np.random.normal(loc=170, scale=10, size=num_samples)  # 平均170cm, 標準偏差10
weight = np.random.normal(loc=65, scale=8, size=num_samples)   # 平均65kg, 標準偏差8

# データフレームを作成
df = pd.DataFrame({
    'Height (cm)': height,
    'Weight (kg)': weight
})

# データの統計情報を表示
print("基本統計情報:")
print(df.describe())

# 散布図を描画
plt.figure(figsize=(8, 5))
plt.scatter(df["Height (cm)"], df["Weight (kg)"], color='blue', alpha=0.5)
plt.xlabel("Height (cm)")
plt.ylabel("Weight (kg)")
plt.title("身長と体重の相関関係")
plt.grid(True)
plt.show()
```
